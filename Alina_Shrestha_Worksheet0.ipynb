**Problem 1: Array Creation**

```python
import numpy as np

# 1. Empty array 2x2
arr1 = np.empty((2,2))
print("Empty Array:\n", arr1)

# 2. All ones array 4x2
arr2 = np.ones((4,2))
print("\nOnes Array:\n", arr2)

# 3. Array filled with specific value
arr3 = np.full((3,3),7)
print("\nFull Array:\n", arr3)

# 4. Zeros array same shape as arr2
arr4 = np.zeros_like(arr2)
print("\nZeros like:\n", arr4)

# 5. Ones array same shape as arr1
arr5 = np.ones_like(arr1)
print("\nOnes like:\n", arr5)

# 6. Convert list to numpy array
new_list = [1,2,3,4]
arr6 = np.array(new_list)
print("\nList to Array:\n", arr6)
```

**Problem 2: Array Manipulation**

```python
# 1. Array from 10 to 49
arr = np.arange(10,50)
print("10 to 49:\n",arr)

# 2. 3x3 matrix from 0 to 8
matrix = np.arange(9).reshape(3,3)
print("\n3x3 Matrix:\n",matrix)

# 3. Identity Matrix
identity = np.eye(3)
print("\nIdentity Matrix:\n",identity)

# 4. Random array size 30 + mean
rand_arr = np.random.random(30)
print("\nRandom Array Mean:", rand_arr.mean())

# 5. 10x10 random array min and max
rand_matrix = np.random.random((10,10))
print("\nMin:", rand_matrix.min())
print("Max:", rand_matrix.max())

# 6. Zero array size 10 replace 5th element
zero_arr = np.zeros(10)
zero_arr[4] = 1
print("\nReplace 5th element:\n",zero_arr)

# 7. Reverse array
arr = np.array([1,2,0,0,4,0])
rev = arr[::-1]
print("\nReversed Array:\n",rev)

# 8. 2D array border 1 inside 0
border = np.ones((5,5))
border[1:-1,1:-1] = 0
print("\nBorder Matrix:\n",border)

# 9. 8x8 checkerboard
checker = np.zeros((8,8))
checker[1::2,::2] = 1
checker[::2,1::2] = 1
print("\nCheckerboard:\n",checker)
```

**Problem 3: Array Operations**

```python
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])

v = np.array([9,10])
w = np.array([11,12])

# 1. Addition
print("Addition:\n", x+y)

# 2. Subtraction
print("\nSubtraction:\n", x-y)

# 3. Multiply with integer
print("\nMultiply with 3:\n", x*3)

# 4. Square each element
print("\nSquare:\n", x**2)

# 5. Dot product
print("\nDot product v.w:", np.dot(v,w))
print("Dot product x.v:", np.dot(x,v))
print("Dot product x.y:\n", np.dot(x,y))

# 6. Concatenate
print("\nConcatenate x and y (rows):\n", np.concatenate((x,y),axis=0))

print("\nConcatenate v and w (columns):\n", np.vstack((v,w)))

# 7. Concatenate x and v
try:
    print(np.concatenate((x,v)))
except Exception as e:
    print("Error:",e)
```

**Problem 4: Matrix Operations**

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
identity = np.dot(A,A_inv)

print("A * A^-1 =\n",identity)

# 2. Prove AB ≠ BA
AB = np.dot(A,B)
BA = np.dot(B,A)

print("\nAB =\n",AB)
print("\nBA =\n",BA)

# 3. Prove (AB)^T = B^T A^T

left = (np.dot(A,B)).T
right = np.dot(B.T,A.T)

print("\n(AB)^T =\n",left)
print("\nB^T A^T =\n",right)
```

**10.2 Experiment: How Fast is NumPy**

```python
import time
import numpy as np
```

**1. Element-wise Addition**

```python
size = 1000000

list1 = list(range(size))
list2 = list(range(size))

start = time.time()

result = [list1[i] + list2[i] for i in range(size)]

print("Python List Addition Time:", time.time()-start)


# NumPy

arr1 = np.array(list1)
arr2 = np.array(list2)

start = time.time()

result = arr1 + arr2

print("NumPy Addition Time:", time.time()-start)
```

**2. Element-wise Multiplication**

```python
start = time.time()

result = [list1[i] * list2[i] for i in range(size)]

print("Python Multiplication Time:", time.time()-start)


start = time.time()

result = arr1 * arr2

print("NumPy Multiplication Time:", time.time()-start)
```

```python
start = time.time()

dot = sum(list1[i]*list2[i] for i in range(size))

print("Python Dot Product Time:", time.time()-start)


start = time.time()

dot = np.dot(arr1,arr2)

print("NumPy Dot Product Time:", time.time()-start)
```
